```python
data = pd.read_csv('kimo.csv')


sleep_hours = data["On average, how many hours of sleep do you get per night during the week (Sunday-Thursday)?"]


sleep_hour_counts = sleep_hours.value_counts().sort_index()


plt.bar(sleep_hour_counts.index, sleep_hour_counts.values)


plt.title("Sleep Hours per Night during the Week")
plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")

# Display the chart
plt.show()
```

```python
df = pd.read_csv("kimo.csv")

column1_values = df['On average, how many hours of sleep do you get per night during the week (Sunday-Thursday)?']
column2_values = df['What is your current Cumulative Grade Point Average (CGPA) on a scale of 0 to 4.0?']


contingency_table = pd.crosstab(column1_values, column2_values)


chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:", expected)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("kimo.csv")

# Assuming you have two categorical columns, for example, "Column1" and "Column2"
# Replace "Column1" and "Column2" with the actual column names from your dataset

column1_values = df['On average, how many hours of sleep do you get per night during the week (Sunday-Thursday)?']
column2_values = df['What is your current Cumulative Grade Point Average (CGPA) on a scale of 0 to 4.0?']

# Create a contingency table
contingency_table = pd.crosstab(column1_values, column2_values)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Chi-squared Test: Heatmap of Contingency Table")
plt.xlabel("What is your current Cumulative Grade Point Average (CGPA) on a scale of 0 to 4.0?")
plt.ylabel("On average, how many hours of sleep do you get per night during the week (Sunday-Thursday)?")
plt.show()
```

```python

df = pd.read_csv("kimo.csv")


column1_values = df['On average, how many hours of sleep do you get per night during the week (Sunday-Thursday)?']
column2_values = df['What is your current Cumulative Grade Point Average (CGPA) on a scale of 0 to 4.0?']


column1_counts = column1_values.value_counts()
column2_counts = column2_values.value_counts()


plt.figure(figsize=(10, 6))

plt.bar(column1_counts.index, column1_counts.values, alpha=0.5, label='On average, how many hours of sleep do you get per night during the week (Sunday-Thursday)?', color='blue')
plt.bar(column2_counts.index, column2_counts.values, alpha=0.5, label='What is your current Cumulative Grade Point Average (CGPA) on a scale of 0 to 4.0?', color='red')

plt.title('Bar Plot of Column 1 and Column 2')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```
